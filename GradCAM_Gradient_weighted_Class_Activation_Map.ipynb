# GradCAM

## **Grad-CAM (Gradient-weighted Class Activation Mapping)**

Grad-CAM (Gradient-weighted Class Activation Mapping) is a popular technique used in deep learning to visualize and interpret the decisions made by convolutional neural networks (CNNs). It helps in understanding which regions of an input image contributed most to the modelâs predictions.

---

### **1. Why is Grad-CAM Needed?**
Deep learning models, especially CNNs, are often considered "black boxes" because they make predictions without explicitly showing why. Grad-CAM provides a way to **interpret** and **visualize** the decision-making process of CNNs, which is crucial for:
- **Debugging Models** â Understanding misclassifications.
- **Model Explainability** â Making AI decisions transparent.
- **Medical Applications** â Highlighting key regions in X-rays, MRIs.
- **Bias Detection** â Checking if a model focuses on irrelevant features.

---

### **2. How Does Grad-CAM Work?**
Grad-CAM generates a heatmap that highlights the **most important regions** in an image that influenced the model's decision. It does this by computing the **gradient of the predicted class score** concerning the final convolutional layerâs feature maps.

### **Step-by-Step Process:**
1. **Forward Pass:**
   - The input image passes through the CNN, and the feature maps are extracted at a specific convolutional layer (usually the last one before fully connected layers).
   - The model makes a prediction and outputs class scores.

2. **Compute Gradients:**
   - The gradient of the **predicted class score** (before softmax) with respect to the feature maps is calculated.
   - These gradients indicate how much each pixel in the feature maps influenced the class prediction.

3. **Compute Importance Weights ($ Î±_k $):**
   - The importance of each feature map is determined by averaging the gradients over spatial dimensions:
     $$
     \alpha_k = \frac{1}{Z} \sum_{i} \sum_{j} \frac{\partial y^c}{\partial A_{ij}^k} $$
     where:
     - $ y^c $ is the class score for class $ c $.
     - $ A^k $ is the $ k^{th} $ feature map.
     - $ Z $ is the total number of pixels in the feature map.

4. **Weighted Combination of Feature Maps:**
   - The feature maps are multiplied by their corresponding weights:
     $$
     L_{\text{Grad-CAM}}^c = \text{ReLU} \left( \sum_k \alpha_k A^k \right) $$
   - ReLU ensures that only **positive** influences are considered (negative values are set to zero).

5. **Generate Heatmap:**
   - The weighted sum produces an attention map.
   - The heatmap is resized to the original image size and superimposed on the input image to highlight important areas.

### **4. Differences Between CAM, Grad-CAM, and Grad-CAM++**
| Method  | Requirement | Advantages | Disadvantages |
|---------|------------|------------|---------------|
| **CAM (Class Activation Mapping)** | Requires model modification (Global Average Pooling) | Directly visualizes activations | Only works for specific architectures |
| **Grad-CAM** | Works with any CNN (no modification) | Uses gradients for better localization | Can be coarse at times |
| **Grad-CAM++** | Uses pixel-wise weighting | More accurate for multiple objects | Computationally more expensive |

1. **`train_test_split(data, test_size=0.2, random_state=42, stratify=data.label.values)`**  
   - This function comes from **scikit-learn's `train_test_split`** module.
   - It **splits** the dataset into:
     - `train_df` â **80%** of the data (training set).
     - `valid_df` â **20%** of the data (validation set).

2. **Parameters Used:**
   - **`data`** â The complete dataset (a Pandas DataFrame).
   - **`test_size=0.2`** â 20% of the data is used for validation.
   - **`random_state=42`** â Ensures the split is **reproducible** (so the same split happens every time the code runs).
   - **`stratify=data.label.values`** â Ensures **class distribution** in both `train_df` and `valid_df` is **preserved**.  
     - This is useful if the dataset is **imbalanced** (some classes have more samples than others).
     - Without `stratify`, the split might randomly assign too many examples of some classes to one set, causing biased training.

### **Why Use Stratified Sampling?**
- If a dataset contains **highly imbalanced** classes (e.g., 80% cats, 20% dogs), a **random split** might put most of the minority class in one set.
- **Stratification ensures that both training and validation sets have the same proportion of each class**.

### **Example Without Stratification:**
If the dataset has:
- **90%** Class A
- **10%** Class B

A random split **might** result in:
- Training Set â 95% Class A, 5% Class B  
- Validation Set â 85% Class A, 15% Class B  

This can lead to poor generalization. Using `stratify`, both sets will have the original **90:10 ratio**.

````python
class ImageModel(nn.Module):
    def __init__(self):
        super(ImageModel, self).__init__()

        self.feature_extractor = nn.Sequential(
            nn.Conv2d(in_channels=3, out_channels=16, kernel_size=(5, 5), padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(4, 4), stride=2),

            nn.Conv2d(in_channels=16, out_channels=16, kernel_size=(5, 5), padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(4, 4), stride=2),

            nn.Conv2d(in_channels=16, out_channels=32, kernel_size=(5, 5), padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(4, 4), stride=2),

            nn.Conv2d(in_channels=32, out_channels=64, kernel_size=(5, 5), padding=1),
            nn.ReLU(),
        )

        self.maxpool = nn.MaxPool2d(kernel_size=(4, 4), stride=2)

        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(6400, 2048),
            nn.ReLU(),
            nn.Linear(2048, 3)
        )

        self.gradients = None

    def activation_hook(self, grad):
        self.gradients = grad

    def forward(self, images):
        x = self.feature_extractor(images)
        x.register_hook(self.activation_hook)
        x = self.maxpool(x)
        x = self.classifier(x)

        return x

    def get_activations_gradient(self):
        return self.gradients

    def get_activations(self, x):
        return self.feature_extractor(x)
````

## **ð¹ How Grad-CAM Works in This Model**
1. **Forward pass:**  
   - Image passes through **feature extractor**.
   - Activations (feature maps) are stored.

2. **Backward pass:**

# Get GradCAM

gradcam.png

````python
def get_gradcam(model, image, label, size):
    label.backward()
    gradients = model.get_activations_gradient()
    pooled_gradients = torch.mean(gradients, dim=[0, 2, 3])
    activations = model.get_activations(image).detach()

    for i in range(activations.shape[1]):
        activations[:, i, :, :] *= pooled_gradients[i]

    heatmap = torch.mean(activations, dim=1).squeeze().cpu()
    heatmap = np.maximum(heatmap, 0)
    heatmap /= torch.max(heatmap)
    heatmap = cv2.resize(heatmap.numpy(), (size, size))

    return heatmap
````

## **Understanding `get_gradcam` Function**  

This function generates a **Grad-CAM (Gradient-weighted Class Activation Mapping)** heatmap to highlight **which regions of an image influenced the model's decision** the most.

---

## **ð¹ Step-by-Step Breakdown**  

- **Inputs**:
  - `model`: Trained PyTorch model.
  - `image`: Input image tensor.
  - `label`: The target class label (used for backpropagation).
  - `size`: The output heatmap size (e.g., `224Ã224` for visualization).

---

### **1ï¸â£ Compute Gradients w.r.t. the Target Class**  

- Calls `.backward()` on the `label` to compute **gradients**.  
- Retrieves the **activation gradients** from the model.  
- These gradients tell us how much **each activation map** contributes to the final prediction.  

---

### **2ï¸â£ Pool the Gradients (Global Average Pooling)**

- Computes the **average gradient** across the width & height of the feature map.  
- This gives a **single importance score** per feature map channel.  

---

### **3ï¸â£ Get Feature Map Activations**

- Extracts **feature maps** before the final classification layer.  
- `.detach()` prevents gradient tracking for efficiency.  

---

### **4ï¸â£ Weight Feature Maps by Gradients**

- Multiplies each activation map by its corresponding **importance score**.  
- Highlights **important features** for the decision.  

---

### **5ï¸â£ Generate Heatmap**

- Averages over all feature maps to create a **single grayscale heatmap**.  
- Moves data to CPU for further processing.  

---

### **6ï¸â£ Normalize & Resize Heatmap**

- **Removes negative values** (ReLU-like operation).  
- **Normalizes heatmap** between `0` and `1`.  
- **Resizes heatmap** to match the original image size (using OpenCV).  

---

## **ð¹ Final Output**
- **Returns** a **heatmap** that highlights regions **most responsible** for the modelâs prediction.

````python
# Function to compute Grad-CAM
def get_gradcam(model, image, label, size):
    """
    Computes the Grad-CAM heatmap.
    """
    # Perform backpropagation on the selected class
    label.backward()

    # Get gradients of the target class w.r.t. feature maps
    gradients = model.get_activations_gradient()

    # Global Average Pooling of gradients across channels
    pooled_gradients = torch.mean(gradients, dim=[0, 2, 3])

    # Get feature activations
    activations = model.get_activations(image).detach()

    # Weight activations by importance (pooled gradients)
    for i in range(activations.shape[1]):
        activations[:, i, :, :] *= pooled_gradients[i]

    # Compute the mean across all feature maps
    heatmap = torch.mean(activations, dim=1).squeeze().cpu()

    # Apply ReLU to remove negative values
    heatmap = np.maximum(heatmap, 0)

    # Normalize heatmap
    heatmap /= torch.max(heatmap)

    # Resize heatmap to match the original image size
    heatmap = cv2.resize(heatmap.numpy(), (size, size))

    return heatmap
````

```python
import torch
import torch.nn.functional as F
import numpy as np
import cv2
from fastapi import FastAPI, UploadFile, File
from torchvision import transforms
from PIL import Image
from io import BytesIO

# Load trained model
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = torch.load("best_model.pt", map_location=DEVICE)
model.eval()

# Image preprocessing
def preprocess_image(image_bytes):
    transform = transforms.Compose([
        transforms.Resize((227, 227)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
    image = Image.open(BytesIO(image_bytes)).convert("RGB")
    return transform(image).unsqueeze(0).to(DEVICE)

# Grad-CAM function
def get_gradcam(model, image):
    image.requires_grad = True
    logits = model(image)
    pred_class = logits.argmax(dim=1)

    logits[:, pred_class].backward()
    gradients = model.get_activations_gradient()
    pooled_gradients = torch.mean(gradients, dim=[0, 2, 3])
    activations = model.get_activations(image).detach()

    for i in range(activations.shape[1]):
        activations[:, i, :, :] *= pooled_gradients[i]

    heatmap = torch.mean(activations, dim=1).squeeze().cpu()
    heatmap = np.maximum(heatmap, 0)
    heatmap /= torch.max(heatmap)
    heatmap = cv2.resize(heatmap.numpy(), (227, 227))
    return heatmap, pred_class.item()

# FastAPI Setup
app = FastAPI()

@app.post("/gradcam/")
async def gradcam_api(file: UploadFile = File(...)):
    image_bytes = await file.read()
    image = preprocess_image(image_bytes)

    heatmap, pred_class = get_gradcam(model, image)

    # Convert heatmap to image
    heatmap_img = (heatmap * 255).astype(np.uint8)
    _, encoded_img = cv2.imencode(".png", heatmap_img)
    return {"predicted_class": pred_class, "heatmap": encoded_img.tobytes()}

# Run the API
if __name__ == "__main__":
    import uvicorn
    uvicorn.run(app, host="0.0.0.0", port=8000)
```
